```
def Cal_ratio(ticker):
  stock = yf.Ticker(ticker)
  status=True
  try:
    df = pd.concat([stock.financials,stock.balance_sheet,stock.cashflow])
    df = df.T
    df = df.fillna(0)
    df = df.replace('-',0)
    df = df.astype(float)
    df['Current ratio']= df['Total Current Assets'] / df['Total Current Liabilities']
    df['Cash ratio'] = df['Cash And Cash Equivalents'] / df['Total Current Liabilities']
    df['Debt ratio'] = df['Total Liabilities']/df['Total Assets']
    df['Operating margin ratio'] = df['Net Income From Continuing Ops'] / df['Total Revenue']
    df['Return on assets ratio'] = df['Net Income Applicable To Common Shares'] / df['Total Assets']
    df['Return on equity ratio'] = df['Net Income Applicable To Common Shares'] / df['Total Stockholder Equity']
    df['Earnings per share ratio'] = df['Net Income Applicable To Common Shares'] / df['Common Stock']
    df['Ticker']=ticker
    ratio_columns = ['Ticker','Current ratio','Cash ratio','Debt ratio','Operating margin ratio',
                    'Return on assets ratio','Return on equity ratio','Earnings per share ratio']
  except:
    status=False
    pass
  if status:
    return df[ratio_columns]
  else:
    return status
  
  
```
